```python
# a. Loading an example dataset using scikit-learn

from sklearn import datasets

from sklearn.model_selection import train_test_split

from sklearn import svm

from sklearn.metrics import accuracy_score, classification_report

import matplotlib.pyplot as plt

import joblib

# Load digits dataset

digits = datasets.load_digits()

# Features and target

X = digits.data      # image data flattened into 64 features

y = digits.target    # digit labels: 0 to 9

print("Dataset shape:", X.shape)

print("Target shape:", y.shape)

print("Classes:", digits.target_names)

# Show one image

plt.imshow(digits.images[0], cmap="gray")

plt.title(f"Digit label: {digits.target[0]}")

plt.axis("off")

plt.show()

# b. Learning and Predicting using sklearn.svm.SVC

# Split data into training and testing set

X_train, X_test, y_train, y_test = train_test_split(

    X, y, test_size=0.25, random_state=42

)

# Create SVM model

model = svm.SVC(gamma=0.001)

# Train the model

model.fit(X_train, y_train)

# Predict test data

y_pred = model.predict(X_test)

# Check accuracy

accuracy = accuracy_score(y_test, y_pred)

print("Model Accuracy:", accuracy)

print("\nClassification Report:")

print(classification_report(y_test, y_pred))

# Predict one image

sample_image = X_test[0].reshape(1, -1)

prediction = model.predict(sample_image)

print("Predicted digit:", prediction[0])

print("Actual digit:", y_test[0])

plt.imshow(X_test[0].reshape(8, 8), cmap="gray")

plt.title(f"Predicted: {prediction[0]} | Actual: {y_test[0]}")

plt.axis("off")

plt.show()

# c. Save a model in scikit-learn

joblib.dump(model, "svm_digit_model.pkl")

print("Model saved successfully as svm_digit_model.pkl")

# Load the saved model again

loaded_model = joblib.load("svm_digit_model.pkl")

# Test loaded model

new_prediction = loaded_model.predict(sample_image)

print("Prediction from loaded model:", new_prediction[0])
```
