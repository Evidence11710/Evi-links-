```python
# a) Import Libraries

import pandas as pd

import matplotlib.pyplot as plt

from sklearn.datasets import load_iris

from sklearn.preprocessing import StandardScaler

from sklearn.decomposition import PCA

# b) Import and Standardize Data

iris = load_iris()

X = iris.data

y = iris.target

target_names = iris.target_names

# Convert to DataFrame

df = pd.DataFrame(X, columns=iris.feature_names)

print("Original Iris Data:")

print(df.head())

# Standardize the data

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)

# c) PCA Execution and Project to 2D

pca = PCA(n_components=2)

X_pca = pca.fit_transform(X_scaled)

# Create PCA DataFrame

pca_df = pd.DataFrame(data=X_pca, columns=["Principal Component 1", "Principal Component 2"])

pca_df["Target"] = y

pca_df["Flower Name"] = pca_df["Target"].map({

    0: "setosa",

    1: "versicolor",

    2: "virginica"

})

print("\nPCA Result:")

print(pca_df.head())

print("\nExplained Variance Ratio:")

print(pca.explained_variance_ratio_)

# Plot PCA Result

plt.figure(figsize=(8, 6))

for target, name in enumerate(target_names):

    plt.scatter(

        pca_df[pca_df["Target"] == target]["Principal Component 1"],

        pca_df[pca_df["Target"] == target]["Principal Component 2"],

        label=name

    )

plt.xlabel("Principal Component 1")

plt.ylabel("Principal Component 2")

plt.title("PCA of Iris Dataset Projected to 2D")

plt.legend()

plt.grid(True)

plt.show()
```
